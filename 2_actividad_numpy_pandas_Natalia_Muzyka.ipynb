## Ejercicio numpy

Imagina que tienes datos de temperatura promedio diaria de una ciudad durante un mes (30 días). Quieres realizar un análisis básico de estos datos usando Numpy para obtener información útil como la temperatura promedio del mes, la temperatura máxima y mínima, y los días en los que la temperatura estuvo por encima de la temperatura promedio.

1. **Crear un array de temperaturas:**

  - Genera un array de Numpy con 30 elementos que representen las temperaturas promedio diarias de la ciudad durante un mes.
  - Que se genere con numeros enteros aleatorios, entre los valores 18 y 29

2. **Calcular estadísticas básicas:**

  - Calcula la temperatura promedio del mes.
  - Encuentra la temperatura máxima y mínima del mes.

3. **Identificar días calurosos:**

  - Identifica los días en los que la temperatura estuvo por encima de la temperatura promedio del mes.

4. **Imprimir resultados:**

  - Imprime la temperatura promedio, máxima y mínima.
  - Imprime los índices (días) en los que la temperatura estuvo por encima del promedio.

```python
# Paso 1: Crear un array de temperaturas para 30 días
temperaturas = np.random.randint(18,29, 30)
temperaturas

```

```python
# Paso 2: Calcular estadísticas básicas
temperatura_promedio = np.mean(temperaturas)
temperatura_maxima = np.max(temperaturas)
temperatura_minima = np.min(temperaturas)
print(f"Temperatura promedio del mes: {temperatura_promedio:.2f}")
print(f"Temperatura máxima del mes: {temperatura_maxima}")
print(f"Temperatura mínima del mes: {temperatura_minima}")
```

```python
# Paso 3: Identificar días calurosos
dias_calurosos = np.where(temperaturas > temperatura_promedio)[0]
# np.where para encontrar los índices de los días en que la temperatura estuvo por encima del promedio

# Paso 4: Imprimir resultados
print(f"Días con temperatura por encima del promedio: {dias_calurosos + 1}")  # Añadir 1 para mostrar días del 1 al 30
```

### Ejercicio filter

1. df_filtrado_1: edad mayor que 30 años, salario menor a 100.000 y género ha de ser female

2. df_filtrado_2: salario menor que 100.000, edad mayor o igual a 45 y ordenar por su edad de mayor a menor

3. df_filtrado_3: que hayan realizado alguna compra, que sean 'male', que tengan más de 30 años pero menos de 50 y que su salario sea mayor a 100.000, ordenados por edad (de menor a mayor) y salario (de mayor a menor)

```python
df = pd.read_csv('/content/car_data.csv')
df.head()
```

```python
# 1. Edad mayor que 30 años, salario menor a 100.000 y género ha de ser female
df_filtrado_1 = df[(df["Age"] > 30) & (df["AnnualSalary"] < 100000) & (df["Gender"] == "Female")]
# comprobamos que se haya filtrado correctamente
print("\n>>> El unico valor en genero ha de ser Female: ")
print(df_filtrado_1.Gender.unique())

print("\n>>> El salario anual maximo ha de ser menor a 100.000: ")
print(df_filtrado_1.AnnualSalary.max())

print("\n>>> La edad minima ha de ser 30 años: ")
print(df_filtrado_1.Age.min())

print("\n>>> El df_filtrado_1:")
df_filtrado_1.head()
```

```python
# 2. Salario menor que 100.000, edad mayor o igual a 45 y ordenar por su edad de mayor a menor
df_filtrado_2 = df[ (df["Age"] >= 45) & (df["AnnualSalary"] < 100000)].sort_values(by="Age", ascending=False)

# comprobamos que se haya filtrado correctamente
print("\n>>> El salario anual maximo ha de ser menor a 100.000: ")
print(df_filtrado_2.AnnualSalary.max())

print("\n>>> La edad minima ha de ser 45 años: ")
print(df_filtrado_2.Age.min())

print("\n>>> El df_filtrado_2:")
df_filtrado_2.head()
```

```python
# 3. Que hayan realizado alguna compra, que sean 'male', que tengan más de 30 años pero menos de 50 y que su salario sea mayor a 100.000
# ordenados por edad y salario
df_filtrado_3 = df[ (df["Purchased"]==1) & (df["Gender"]=="Male") & (df["Age"] > 30) & (df["Age"]<50) & (df["AnnualSalary"] > 100000)] \
                .sort_values(by=["Age","AnnualSalary"], ascending=[True, False])

# comprobamos que se haya filtrado correctamente
print("\n>>> El unico valor en purchased ha de ser 1: ")
print(df_filtrado_3.Purchased.unique())

print("\n>>> El unico valor en genero ha de ser Male: ")
print(df_filtrado_3.Gender.unique())

print("\n>>> El salario anual mínimo ha de ser mayor a 100.000: ")
print(df_filtrado_3.AnnualSalary.min())

print("\n>>> La edad minima ha de ser 30 años: ")
print(df_filtrado_3.Age.min())

print("\n>>> La edad máxima ha de ser 50 años: ")
print(df_filtrado_3.Age.max())

print("\n>>> El df_filtrado_3:")
df_filtrado_3.head()
```

### Ejercicio group by

1. Agrupar por `Product` y `Region`, y obtener la suma de las ventas
2. Filtrar el df para incluir solo aquellos productos donde las ventas son mayores que 1000, agrupar por `Product`, y obtener el promedio de las ventas para cada producto
3. Agrupar por `Region`, y obtener la suma de las ventas y el número total de ventas
4. Filtrar el df para incluir solo aquellas regiones donde la suma de ventas es mayor que 2000, agrupar por `Region`, y obtener la suma y el promedio de las ventas para cada región (usar lambda!!!)

```python
# Crear el df de ejemplo
data = {
    'Product': ['A', 'B', 'A', 'C', 'B', 'C', 'A', 'B', 'C', 'A'],
    'Region': ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'],
    'Sales': [1000, 1500, 800, 1200, 900, 1100, 750, 1300, 850, 950]
}

df = pd.DataFrame(data)
df
```

```python
# 1: Agrupar por `Product` y `Region`, y obtener la suma de las ventas.
sales_sum_product_region = df.groupby(['Product', 'Region'])['Sales'].sum().reset_index()
sales_sum_product_region
```

```python
# 2: Filtrar el df para incluir solo aquellos productos donde las ventas son mayores que 1000,
# agrupar por `Product`, y obtener el promedio de las ventas para cada producto.
filtered_df = df[df['Sales'] > 1000]  # Filtrar para ventas mayores que 1000
filtered_df
```

```python
average_sales_product = filtered_df.groupby('Product')['Sales'].mean().reset_index()  # Calcular promedio de ventas por producto
average_sales_product
```

```python
# 3: Agrupar por `Region`, y obtener la suma de las ventas y el número total de ventas.
total_sales_region = df.groupby('Region')['Sales'].agg(['sum', 'count']).reset_index()  # Calcular suma y contar ventas por región
total_sales_region
```

```python
# 4: Filtrar el df para incluir solo aquellas regiones donde la suma de ventas es mayor que 2000,
# agrupar por `Region`, y obtener la suma y el promedio de las ventas para cada región.
# Filtrar regiones con suma de ventas mayor que 2000
filtered_df = df.groupby('Region').filter(lambda x: x['Sales'].sum() > 2000)
filtered_df.sort_values(by="Region")

# Calcular suma y promedio de ventas por región
sales_stats_region = filtered_df.groupby('Region')['Sales'].agg(['sum', 'mean']).round(2).reset_index()
sales_stats_region
```
